```python
import pandas as pd
df = pd.read_csv("data.csv")
df.head()
```

```python
import altair as alt

alt.Chart(df).mark_circle(
    size=300,
    opacity=0.75
).encode(
    x=alt.X('life_expectancy').scale(domain=[74, 82]).axis(domain=False,title=None, ticks=False, values=list(range(75, 82, 1))),
    color=alt.Color('region', legend=None).scale(scheme='pastel1')
).properties(
    height=40,
    width=800
).configure_view(
    strokeWidth=0
)
```
